# K-Means Basic Clustering

## Objective
Understand basic clustering using K-Means.

Tasks:
- Apply K-Means with:
  - K = 2
  - K = 3
  - K = 4
- Plot clusters and centroids
- Analyze clustering behavior


## Step 1: Import Required Libraries

```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
```

## Step 2: Load Dataset

```python
data = pd.read_csv('kmeans_basic.csv')

data.head()
```

## Step 3: Understand Dataset

```python
data.info()
```

## Step 4: Select Features

```python
X = data.iloc[:, :]

print(X.shape)

X.head()
```

## Step 5: Standardize Features

Standardization Formula:

z = (x - mean) / standard deviation


```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

# K-Means Clustering with K = 2

K-Means tries to divide data into 2 groups.


```python
kmeans_2 = KMeans(
    n_clusters=2,
    random_state=42
)

clusters_2 = kmeans_2.fit_predict(X_scaled)

centroids_2 = kmeans_2.cluster_centers_

print(centroids_2)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X_scaled[:,0],
    X_scaled[:,1],
    c=clusters_2,
    cmap='viridis'
)

plt.scatter(
    centroids_2[:,0],
    centroids_2[:,1],
    s=300,
    c='red',
    marker='X'
)

plt.title("K-Means Clustering (K=2)")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```

# K-Means Clustering with K = 3

Now the algorithm divides the data into 3 groups.


```python
kmeans_3 = KMeans(
    n_clusters=3,
    random_state=42
)

clusters_3 = kmeans_3.fit_predict(X_scaled)

centroids_3 = kmeans_3.cluster_centers_

print(centroids_3)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X_scaled[:,0],
    X_scaled[:,1],
    c=clusters_3,
    cmap='viridis'
)

plt.scatter(
    centroids_3[:,0],
    centroids_3[:,1],
    s=300,
    c='red',
    marker='X'
)

plt.title("K-Means Clustering (K=3)")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```

# K-Means Clustering with K = 4

Now the dataset is divided into 4 clusters.


```python
kmeans_4 = KMeans(
    n_clusters=4,
    random_state=42
)

clusters_4 = kmeans_4.fit_predict(X_scaled)

centroids_4 = kmeans_4.cluster_centers_

print(centroids_4)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X_scaled[:,0],
    X_scaled[:,1],
    c=clusters_4,
    cmap='viridis'
)

plt.scatter(
    centroids_4[:,0],
    centroids_4[:,1],
    s=300,
    c='red',
    marker='X'
)

plt.title("K-Means Clustering (K=4)")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```

## Step 6: Understanding K-Means Objective

K-Means minimizes the distance between points and centroids.

Distance Formula:

distance = sqrt((x2-x1)^2 + (y2-y1)^2)


## Step 7: Understanding Centroids

Centroids are the center points of clusters.

They represent:
- Average position of all points inside a cluster
- Cluster prototype


# Final Conclusions

## Q1. How does cluster structure change as K increases?

As K increases:
- More clusters are formed
- Clusters become smaller
- Separation becomes more detailed

Very high K may over-segment the data.

## Q2. Which K gives the most meaningful grouping? Why?

The best K is the one where:
- Clusters are clearly separated
- Centroids represent meaningful groups
- Overlapping is minimized

Usually K=3 works well for naturally separated datasets.

## Q3. What do centroids represent in this dataset?

Centroids represent:
- The average location of each cluster
- The central pattern of data points inside the cluster
