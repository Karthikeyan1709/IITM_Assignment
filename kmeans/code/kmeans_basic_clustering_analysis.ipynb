
# K-Means Clustering Analysis - kmeans_basic.csv

## Objective
Understand the basics of clustering using the K-Means algorithm.

## Tasks
- Apply K-Means with:
  - K = 2
  - K = 3
  - K = 4
- Plot clusters and centroids

## Questions
1. How does cluster structure change as K increases?
2. Which K gives the most meaningful grouping? Why?
3. What do centroids represent in this dataset?


```python

# Import required libraries

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```


# Load Dataset


```python

# Load dataset

df = pd.read_csv('kmeans_basic.csv')

# Display first 5 rows
df.head()
```


# Select Features

We will use all columns for clustering.


```python

X = df.iloc[:, :]

print(X.head())
```


# K-Means with K = 2


```python

kmeans_2 = KMeans(n_clusters=2, random_state=42)

clusters_2 = kmeans_2.fit_predict(X)

centroids_2 = kmeans_2.cluster_centers_

print("Centroids for K=2")
print(centroids_2)

print("Inertia for K=2:", kmeans_2.inertia_)
```

```python

plt.figure(figsize=(8,6))

plt.scatter(X.iloc[:,0], X.iloc[:,1], c=clusters_2)

plt.scatter(
    centroids_2[:,0],
    centroids_2[:,1],
    marker='X',
    s=300
)

plt.title("K-Means Clustering (K=2)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```


# K-Means with K = 3


```python

kmeans_3 = KMeans(n_clusters=3, random_state=42)

clusters_3 = kmeans_3.fit_predict(X)

centroids_3 = kmeans_3.cluster_centers_

print("Centroids for K=3")
print(centroids_3)

print("Inertia for K=3:", kmeans_3.inertia_)
```

```python

plt.figure(figsize=(8,6))

plt.scatter(X.iloc[:,0], X.iloc[:,1], c=clusters_3)

plt.scatter(
    centroids_3[:,0],
    centroids_3[:,1],
    marker='X',
    s=300
)

plt.title("K-Means Clustering (K=3)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```


# K-Means with K = 4


```python

kmeans_4 = KMeans(n_clusters=4, random_state=42)

clusters_4 = kmeans_4.fit_predict(X)

centroids_4 = kmeans_4.cluster_centers_

print("Centroids for K=4")
print(centroids_4)

print("Inertia for K=4:", kmeans_4.inertia_)
```

```python

plt.figure(figsize=(8,6))

plt.scatter(X.iloc[:,0], X.iloc[:,1], c=clusters_4)

plt.scatter(
    centroids_4[:,0],
    centroids_4[:,1],
    marker='X',
    s=300
)

plt.title("K-Means Clustering (K=4)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.show()
```


# Observations

## How does cluster structure change as K increases?

- As K increases, clusters become smaller and more detailed.
- K=2 creates broad clusters.
- K=3 creates better natural grouping.
- K=4 splits one cluster into smaller groups.

---

## Which K gives the most meaningful grouping?

K=3 gives the most meaningful grouping because:
- clusters are clearly separated
- clusters are compact
- data naturally appears to contain 3 groups

---

## What do centroids represent?

Centroids represent:
- the center of each cluster
- the average position of all points in a cluster

Formula:

Centroid = (sum of all points) / number of points
