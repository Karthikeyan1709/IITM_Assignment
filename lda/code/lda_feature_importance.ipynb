# LDA Feature Importance Analysis

## Objective
Interpret LDA coefficients and rank feature importance.

## Step 1: Import Required Libraries

```python
import pandas as pd
import numpy as np

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler
```

## Step 2: Load Dataset

```python
data = pd.read_csv('lda_feature_importance.csv')

print(data.head())
```

## Step 3: Understand Dataset

```python
data.info()
```

```python
data.iloc[:, -1].value_counts()
```

## Step 4: Separate Features and Target

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

print("Feature Shape :", X.shape)
print("Target Shape :", y.shape)
```

## Step 5: Standardize Features

Standardization formula:

z = (x - mean) / standard deviation


```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

## Step 6: Fit LDA Model

```python
lda = LinearDiscriminantAnalysis()

lda.fit(X_scaled, y)
```

## Step 7: Extract LDA Coefficients

```python
coefficients = lda.coef_[0]

print(coefficients)
```

## Step 8: Create Feature Importance Table

```python
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': coefficients
})

feature_importance['Absolute_Coefficient'] = np.abs(feature_importance['Coefficient'])

feature_importance = feature_importance.sort_values(
    by='Absolute_Coefficient',
    ascending=False
)

feature_importance
```

## Step 9: Print Ranked Features

```python
print("Feature Ranking:\n")

for index, row in feature_importance.iterrows():
    print(f"{row['Feature']} --> {row['Absolute_Coefficient']:.4f}")
```

## Step 10: Identify Negligible Features

```python
threshold = 0.1

negligible = feature_importance[
    feature_importance['Absolute_Coefficient'] < threshold
]

print(negligible)
```

LDA Decision Equation:

# Final Conclusions

## Q1. Which features are most important?
Features with the largest absolute coefficients.

## Q2. Are some features negligible?
Yes. Features near zero contribute very little.

## Q3. How do coefficients relate to separation?
Large coefficients increase separation influence.

## Q4. Can LDA be used for feature selection?
Yes. High coefficient features are useful while near-zero coefficient features may be removed.
